```python
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Mumbai'
website_url = requests.get(url).text

soup = BeautifulSoup(website_url,'html.parser')
# print(soup.prettify())
table=soup.find('table')
table
res = []
for tr in table.find_all('tr'):
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)
res

Mumbai_neighborhood = pd.DataFrame(res, columns=["Area", "Location", "Latitude", "Longitude"])
Mumbai_neighborhood
```

```python
address = 'Mumbai, India'

geolocator = Nominatim(user_agent="india_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Mumbai, India are {}, {}.'.format(latitude, longitude))
```

```python
map_mumbai = folium.Map(location=[latitude, longitude], zoom_start=12)

# add markers to map
for lat, lng, area, location in zip(Mumbai_neighborhood['Latitude'], Mumbai_neighborhood['Longitude'], Mumbai_neighborhood['Area'], Mumbai_neighborhood['Location']):
    label = '{}, {}'.format(location, area)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_mumbai)  
    
map_mumbai
```

```python
url='https://memumbai.com/ward-wise-breakdown-of-covid-19-positive-cases/'

covid_data = pd.read_html(url)
temp_df = covid_data[0]
temp_df.drop([0],axis=0,inplace=True)
temp_df.rename(columns={0:"Ward",1:"Discharged",2:"Deaths",3:"Active",4:"Total Patients"},inplace=True)
temp_df = temp_df.astype({'Ward':str,'Discharged':str,'Deaths':str,'Active':str,'Total Patients':str})
covid_data_df = pd.DataFrame({'Ward':temp_df.loc[1]['Ward'].split(' '),'Discharged':temp_df.loc[1]['Discharged'].split(' '),'Deaths':temp_df.loc[1]['Deaths'].split(' '),'Active':temp_df.loc[1]['Active'].split(' '),'Total Patients':temp_df.loc[1]['Total Patients'].split(' ')})
covid_data_df = covid_data_df.astype({'Ward':str,'Discharged':int,'Deaths':int,'Active':int,'Total Patients':int})
covid_data_df
```

```python
mumbai_ward_details = pd.read_csv('Mumbai_ward_details.csv')
mumbai_ward_details
```

```python
Final_covid_data_df = pd.merge(mumbai_ward_details,covid_data_df,on='Ward')
Final_covid_data_df
```

```python
df = Final_covid_data_df[['Ward','Total Patients']]
df=df.sort_values("Total Patients",ascending=False)

df.set_index('Ward',inplace = True)

ax = df.plot(kind='bar', figsize=(10, 6), rot=0)

ax.set_ylabel('Number of Total COVID Patients')
ax.set_xlabel('Ward')
ax.set_title('Neighbourhoods in Mumbai with the number of Covid patients')

for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14,
               )

plt.show()
```
